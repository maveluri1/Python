```python
# Finding best out of bag (oob) score Vs no of trees
#random_forest = RandomForestClassifier(oob_score=True)

estimators = np.arange(10, 300, 10)
scores = []
for n in estimators:
    random_forest.set_params(n_estimators=n)
    random_forest.fit(X_train, y_train)
    scores.append(random_forest.oob_score_)
    
df_scores=pd.DataFrame({'Scores': scores, 'Number of Tress': estimators})
print(df_scores)
plt.title("Effect of n_estimators")
plt.xlabel("n_estimators")
plt.ylabel("score")
plt.plot(estimators, scores)
```
